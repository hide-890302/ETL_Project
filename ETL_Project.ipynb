### Key
TEC - Total Energy Consumption
<br>
REC - Renewable Energy Consumption
<br>
ktoe - Tonne of oil equivalent

```python
TEC_df = pd.read_csv('Resources/Total_Final_Renewable_energy_consumption_clean.csv')
TEC_df.columns.values[0] = "Country Name"
TEC_df.head()
```

```python
REC_df = pd.read_csv('Resources/Renewable_energy_consumption_clean.csv')
REC_df.head()
```

```python
# Merge two dataframes using an inner join and set suffixes
TEC_REC_merge_df = pd.merge(TEC_df, REC_df, 
                    on="Country Name", suffixes=("_TEC_ktoe", "_REC_PCT"))
TEC_REC_merge_df.head()
```

```python
# Save Dataframe to .csv
TEC_REC_merge_df.to_csv("Resources/TEC_REC_merge.csv")
```
